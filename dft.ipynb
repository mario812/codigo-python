<h4> CALCULO DE LA DFT </h4>
En este script se calcula la DFT (transformada discreta de Fourier) de una señal
compuesta de la superposición  de 3 sinusoides de amplitudes y frecuencias diferentes.
El tiempo discreto consiste de 100 valores igualmente espaciados

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_muestras = 100
interval = 1 / num_muestras

# Generación del tiempo discreto
t_start  = 0
t_end = 1
t = np.arange(t_start, t_end, interval)
t
```

A continuación se genera la señal del tiempo discreto <code>X_t</code> y se grafica.

```python

v1 = 2.3 * np.sin(2*np.pi * 10 * t)
v2 = 10 * np.sin(2*np.pi * 1 * t)
v3 = 9 * np.sin(2*np.pi * 2 * t)
X_t = v1 + v2 + v3
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, X_t, 'r')
plt.xlabel('t')
plt.ylabel('x')
plt.grid(True)
```

```python
# se define un vector de índices para manejar las sumatorias
N = len(X_t)
n = np.arange(N) 
n
```

```python
# cálculo de la DFT
c1 = 2 * np.pi/N
X_k = np.zeros_like(n, dtype=complex)
for r in n:
    for col in n:
        X_k[r] += X_t[col] * np.exp(-1j * c1 * r * col)
```

```python
X_k
```

```python
T = N / num_muestras

freq = n / T
print(freq)
plt.figure(figsize=(8,6))
plt.stem(freq, abs(X_k), 'b', markerfmt=' ', basefmt='-b')
plt.xlabel(r'$\omega$ '+'(Hz)')
plt.ylabel('DFT Amplitude '+r'$|X(\omega)|$')
plt.grid(True)
```
